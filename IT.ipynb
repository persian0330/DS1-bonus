Outlier outlier: refers to individual values in a sample whose values deviate significantly from the rest of the observations of the sample to which it (or they) belong, also known as anomalous data, outliers. When encountering a small number of outboundrs in a set of data, it is generally necessary to reject them to avoid interfering with the correct results. 


We can detect and remove outbounds through boxplots.


The boxplot can be summarized by 5 numbers: minimum, maximum, lower quartile (Q1, at 25%), upper quartile (Q3, at 75%), median (50% position). 


Boxplots can show outliers and can be calculated using IQR (InterQuartile Range, quartile distance Q3-Q1). This outlier is defined as an outlier defined as a value less than Q1-1.5IQR or greater than Q3+1.5IQR.


UpperLimit=Q3+1.5IQR=75% quantile +1.5 (75% quantile - 25% quantile)


LowerLimit=Q1-1.5IQR=25% quantile - 1.5 (75% quantile -25% quantile)


That is, values other than UpperLimit and LowerLimit are outbound.


## reference:http://events.jianshu.io/p/9e02b33b6217

We can see that all the figures have outliers except company size. So we must get remove those outliers.
Thus we define a function to determine the outliers and remove them

```python
#determine and remove the outliers
def Outlier_detection(dataframe, i):
    Q1 = dataframe[i].quantile(0.25)
    Q3 = dataframe[i].quantile(0.75)
    IQR = Q3 - Q1
    low = Q1 - 1.5 * IQR
    high = Q3 + 1.5 * IQR
    df_new = dataframe.loc[(dataframe[i] >= low) & (dataframe[i] <= high)]
    return df_new
```

```python
df_age = Outlier_detection(df, 'Age')
df_experience = Outlier_detection(df_age, 'Total years of experience')
df_salary = Outlier_detection(df_experience, 'Yearly brutto salary (without bonus and stocks) in EUR')
df_bonus = Outlier_detection(df_salary,'Yearly bonus + stocks in EUR')
df_vacation = Outlier_detection(df_bonus, 'Number of vacation days')
df=df_vacation
```

### 3 Hypothesis Test

```python
df.info()
```

```python
df['Position '].value_counts()
```

The two most populated positions are Software Engineer and Backend Developer.

```python
df_SE = df.loc[df['Position ']=='Software Engineer']#Sofware Engineer
df_BD = df.loc[df['Position ']=='Backend Developer']#Backend Developer
```

```python
df_SE.describe().T
```

```python
df_BD.describe().T
```

According to the previous EDA, it seems like that the Yearly brutto salary (without bonus and stocks) of Software Engineer and Backend Developer are almost same. So it is hypothesized that they are the same, but we should first test the variance equality.

$H_0: \sigma_{SE}^2 = \sigma_{BD}^2 $

$H_1: \sigma_{SE}^2 \neq \sigma_{BD}^2 $

```python
df_SE_var = df_SE.loc[:,'Yearly brutto salary (without bonus and stocks) in EUR'].var()
df_BD_var = df_BD.loc[:,'Yearly brutto salary (without bonus and stocks) in EUR'].var()
print(df_SE_var)
print(df_BD_var)
```

```python
import scipy#for hypothesis testing
f = df_SE_var/df_BD_var
p_value = 1 - scipy.stats.f.cdf(f, len(df_SE)-1, len(df_BD)-1)
print(f)
print(p_value)
```

We assume that Level of significance is 0.05. Since 0.4779282744613369>0.05, we accept $H_0$: The variance of two sets are not siginificantly different. 

For the two-sample case with equal variance we choose t-test and we hypothesized $H_0$ and $H_1$.

$H_0: \mu_{SE} - \mu_{BD} = 0$

$H_1: \mu_{SE} - \mu_{BD} ≠ 0$

```python
df_SE_YBS = df_SE['Yearly brutto salary (without bonus and stocks) in EUR']
df_BD_YBS = df_BD['Yearly brutto salary (without bonus and stocks) in EUR']
scipy.stats.ttest_ind(df_SE_YBS, df_BD_YBS, equal_var=True)
```

Since 0.5292654493462899>0.05, we accept $H_0$: The Yearly brutto salary (without bonus and stocks) of Software Engineer and Backend Developer are not significantly different.

### 4 Feature Extraction

Feature Extraction is to transform/compress data in order to reduce dimensionality.

Having previously dealt with missing values and outliers, the features of the categorical data now need to be extracted for later model learning.

The random forest classifier will be used in the model training, which can handle data with high dimensionality (many features) and without dimensionality reduction, without doing feature selection, and can also determine the importance of features and can determine the interaction between different features. Therefore, there is no need to do feature selection and only need to encode the categorical data.

```python
#categorical data
categorical_data = [i for i in df.columns if df[i].dtype == 'object']

#remove the feature that we don't need
del categorical_data[5]
df[categorical_data]
```

```python
import category_encoders as ce #categorical data encoding
```

```python
catagorical_encoder = ce.OneHotEncoder(cols=['Gender', 'City', 'Position ', 'Seniority level', 'Your main technology / programming language',  'Employment status', 'Сontract duration', 'Main language at work', 'Company type'],  
                        use_cat_names=True).fit(df[categorical_data])
catagorical_encoded = catagorical_encoder.transform(df[categorical_data]) # convert the dataset to one-hot
```

### 5 Statistical Learning

Get y (response variable) from 'Yearly brutto salary' column

```python
y = data.iloc[:,data.shape[1]-1]
x = data.iloc[:,range(0,data.shape[1]-1)]
```

Randomly split dataset. 80% of the dataset is used to train the model, 20% is used to test the model.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
x_train
```

```python
y_train
```

Training random forest classifier.
This model is based on Gini Impurity:
$$Gini = 1 - \sum_{C}^{i=1}(p_{i}^2) $$
$p_{i}$ is the probability that a data point is exactly in the $i-$th class is $p_{i}$. \\
This means that smaller the Gini Impurity, the higher the purity of the class and the better the classifier performs.

```python
from sklearn.ensemble import RandomForestClassifier
randomForest_10 = RandomForestClassifier(n_estimators = 10,random_state = 0)
randomForest_10.fit(x_train,y_train)
predictions = randomForest_10.predict(x_test)

from sklearn.metrics import accuracy_score
print('Model accuracy score of random forest with 10 trees : {0:0.4f}'. format(accuracy_score(y_test, predictions)))
```

Changing the hyperparameter (the number of trees) from 10-50, we found that the accuracy is stable.

```python
randomForest_50 = RandomForestClassifier(n_estimators = 50,random_state = 0)
randomForest_50.fit(x_train,y_train)
predictions = randomForest_50.predict(x_test)

print('Model accuracy score of random forest with 50 trees : {0:0.4f}'. format(accuracy_score(y_test, predictions)))
```

### 6 Hypothesis Testing of Learning Methods

Explore whether the number of trees has influence on the accuracy of the models.

$H_0: \mu_1 - \mu_2 = 0$

$H_1: \mu_1 - \mu_2 ≠ 0$

```python
from mlxtend.evaluate import paired_ttest_5x2cv
t, p = paired_ttest_5x2cv(estimator1 = randomForest_10, 
                          estimator2 = randomForest_50, 
                          X=x_train, 
                          y=y_train, 
                          scoring='accuracy', 
                          random_seed=1)
print(f'The P-value is = {p:.4f}')
print(f'The t-statistics is = {t:.4f}')
if p <= 0.05:
    print('Since p<0.05, We can reject the null-hypothesis that and conclude that the performance these 2 forests are significantly different.')
else:
    print('Since p>0.05, we cannot reject the null-hypothesis and can conclude that the performance of these 2 forests are not significantly different.')
```
